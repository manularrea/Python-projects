### 5. Integración e imputación de missing values.

```python
def impute_missing_values(df, dfnull, dfhelper):
    '''
    Imputa valores faltantes en las columnas 'height_m' y 'weight_kg' de df utilizando los valores correspondientes 
    de dfhelper en base al 'pokedex_number'.

  Args:
    df (dataframe): Un Pandas DataFrame con valores faltantes.
    dfnull (dataframe): Un Pandas DataFrame que contiene los valores faltantes.
    dfhelper (dataframe): Un Pandas DataFrame que contiene los valores de referencia.

  Returns:
    Un Pandas DataFrame con los valores faltantes imputados.
    '''
    
    pokedex_null = dfnull['pokedex_number'].tolist()

    for index, row in df.iterrows():
        if row['pokedex_number'] in pokedex_null:
            helper_row = dfhelper[dfhelper['pokedex_number'] == row['pokedex_number']]
            df.at[index, 'height_m'] = helper_row['height_m'].values[0]
            df.at[index, 'weight_kg'] = helper_row['weight_kg'].values[0]

    return df
```

```python
df = impute_missing_values(df, dfnull, dfhelper)
```

```python
# Se verifica nuevamente si hay valores nulos.
df.isnull().values.any()
```

#### 4.3 ¿Qué tipo de Pokémon es más probable que sea un Pokémon legendario?

```python
# Probabilidad de que un pokémon sea legendario segun el Tipo 1
legendary_counts = df[df['is_legendary'] == 1]['type1'].value_counts()
print((legendary_counts/len(df[df['is_legendary'] == 1]))*100)
```

```python
# Probabilidad de que un pokémon sea legendario segun el Tipo 2
legendary_counts = df[df['is_legendary'] == 1]['type2'].value_counts()
exclude_type = 'HHH'
legendary_counts_filtered = legendary_counts[legendary_counts.index != exclude_type]
print((legendary_counts_filtered/len(df[df['is_legendary'] == 1]))*100)
```

El tipo secundario de Pokémon Flying es el más probable que sea legendario, con una probabilidad de 14.3%
